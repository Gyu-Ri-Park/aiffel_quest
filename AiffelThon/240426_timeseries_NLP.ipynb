- 지역: "NWE", "Med", "Los Angeles", "Singapore", "Rotterdam" 등
- 경제/시장 용어: "(NextGen MOC)", "ICE", "Settlement", "Mo01", "Mo01", "Barge"
    - NextGen MOC: 이는 특정 거래 플랫폼이나 거래 방식을 나타낼 수 있으며, 표준 거래 방식과 다를 수 있음.
    - ICE (Intercontinental Exchange): 글로벌 시장에서 에너지, 금융 상품 등을 거래하는 교환 시장
    - Settlement: 거래가 종료된 후 최종적으로 정산된 가격
    - Mo01, Mo02: 해당 거래의 월별 계약. Mo01은 첫 번째 월 계약, Mo02는 두 번째 월 계약을 의미.
    - Barge: 운송 수단이 바지선인 경우.

전부 황함량이 0.1%이므로 'Low Sulfur'라는 키워드는 삭제한다.

```python
data['Assessment'] = data['Assessment'].str.replace('Low Sulfur', '', regex=False)
data['Assessment'].str.contains('Low Sulfur').any()
```

```python
def classify_fuel_type(value):
    if "Marine Gasoil" in value:
        return "Marine Gasoil"
    elif "Marine Fuel" in value:
        return "Marine Fuel"
    elif "Marine Diesel" in value:
        return "Marine Diesel"
    elif "Gasoil" in value:
        return "Gasoil"
    elif "Bunker FO" in value:
        return "Bunker Fuel Oil"
    elif "FO" in value:
        return "Fuel Oil"
    else:
        return "Other"

def classify_sulfur_content(value):
    if ".001%S (10ppm)" in value:
        return "0.001%"
    elif "0.1%S" in value:
        return "0.1%"
    elif "0.1%" in value:
        return "0.1%"
    elif ".1%S (1000ppm)" in value:
        return "0.1%"
    elif "3.5S%" in value:
        return "3.5%"
    elif "3.5%" in value:
        return "3.5%"
    elif "0.5%" in value:
        return "0.5%"
    else:
        return "Unknown"

def classify_viscosity(value):
    if "180 CST" in value:
        return "180 CST"
    elif "380 CST" in value:
        return "380 CST"
    elif "500 CST" in value:
        return "500 CST"
    else:
        return "Unknown"

def classify_delivery_type(value):
    if "Dlvd" in value:
        return "Delivered"
    elif "Bunkers Delivered" in value:
        return "Delivered"
    elif "Delivered" in value:
        return "Delivered"
    elif "FOB" in value:
        return "FOB"
    elif "CIF" in value:
        return "CIF"
    elif "Ex-Wharf" in value:
        return "Ex-Wharf"
    else:
        return "Unknown"

def classify_others(value):
    if "(NextGen MOC)" in value:
        return "NextGen MOC"
    elif "ICE" in value:
        return "ICE"
    elif "Settlement Mo01" in value:
        return "Settlement Mo01"
    elif "Settlement Mo02" in value:
        return "Settlement Mo02"
    elif "Barge" in value:
        return "Barge"
    else:
        return "None"

# 'etc' 컬럼을 위한 데이터 추출 함수
def extract_etc(value, fuel_type, sulfur_content, viscosity, delivery_type, others):
    # 사용된 분류를 제외한 나머지 텍스트 추출
    etc_text = value.replace(fuel_type, '').replace(sulfur_content, '').replace(viscosity, '').replace(delivery_type, '').replace(others, '')
    return etc_text.strip()
```

```python
# 각 데이터 포인트를 분류
data['Fuel Type'] = data['Assessment'].apply(classify_fuel_type)
data['Sulfur Content'] = data['Assessment'].apply(classify_sulfur_content)
data['Viscosity'] = data['Assessment'].apply(classify_viscosity)
data['Delivery Type'] = data['Assessment'].apply(classify_delivery_type)
data['Others'] = data['Assessment'].apply(classify_others)

# 'etc' 컬럼 계산
data['Etc'] = data.apply(lambda x: extract_etc(x['Assessment'], x['Fuel Type'], x['Sulfur Content'], x['Viscosity'], x['Delivery Type'], x['Others']), axis=1)

data
```

```python
data.rename(columns = {'Etc': 'Location'}, inplace = True)
keywords_to_remove = ['Bunker Fo', 'FO', 'S ', 'Dlvd', '.001%S (10ppm) ', '.1%S (1000ppm) ', '.1% (1000ppm)', '.1%(1000ppm)', '.001%(10ppm)',  '.001%(10ppm) ', 
                      'Bunkers', 'Bunker ', '()', ' $']

def remove_keywords(text):
    for keyword in keywords_to_remove:
        text = text.replace(keyword, '')
    return text.strip()

data['Location'] = data['Location'].apply(remove_keywords)
data
```

```python
data['Statistics'] = data['Statistics'].str.replace('mt/CLOSE', 'CLOSE', regex = False)

replacement = {'Settlement Mo01': 'Singapore', 'Settlement Mo02': 'Singapore', 'Spore Cargo': 'Singapore', 'Rdam': 'Rotterdam', 'Tokyo Bay': 'Japan'}
data['Location'] = data['Location'].replace(replacement)

data
```
